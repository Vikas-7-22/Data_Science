```python
import pandas as pd
d={'one':pd.Series([1,2,3],index=['a','b','c']),
   'two':pd.Series([1,2,3,4],index=['a','b','c','d'])
  }
df=pd.DataFrame(d)
print("Adding a new column by passing as Series:")
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)
print("Adding a new column using the existing columns in DataFrame:")
df['four']=df['one']+df['three']
print(df)
```

```python
import pandas as pd
d={'one':pd.Series([1,2,3],index=['a','b','c']),
   'two':pd.Series([1,2,3,4],index=['a','b','c','d'])
  }
df=pd.DataFrame(d)
print("Our dataframe is:")
print(df)
print("Deleting the first column using DEL function:")
del df['one']
print(df)
print("Deleting another column using POP function:")
df.pop('two')
print(df)
print(df[1:4])
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'budget':[1000,2000,np.nan,10,1000]
})
df['budget'].fillna(df['budget'].mode()[0],inplace=True)
print(df)
```

```python
df_ffill=df.copy()
df_ffill.fillna(method='ffill',inplace=True)
print(df_ffill)
```

```python
df_bfill=df.copy()
df_bfill.fillna(method='bfill',inplace=True)
print(df_bfill)
```

```python
import pandas as pd

# Sample dataset with inconsistent date formats
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

# Convert all dates to standard ISO format (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
df= pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID']
})

# Convert all names to lowercase
df['Name_lower'] = df['Name'].str.lower()
print(df)

# Convert all names to uppercase
df['Name_upper'] = df['Name'].str.upper()

print(df.head(10))
```
